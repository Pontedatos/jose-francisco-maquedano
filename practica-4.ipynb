Ahora vamos a hacer un gráfico de barras para alguno de los medios que se encuentran en la tabla. Como los datos de la tabla no están previstos para analizarlos de esta manera, tendremos que introducir los datos manualmente del gráfico. Lo haremos doble para comparar los contratos percibidos por un medio o por otro en los distintos años. En este caso vamos a comparar, Aragon Radio y El Periodico de Aragon que acabamos de buscar

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
años = ["2011", "2012", "2013", "2014", "2015", "2016", "2017"]
aragon_radio = [4962, 6306, 6306, 6306, 4413, 5328, 0]
aragon_periodico = [41972, 38475, 62613, 43932, 40313, 31710, 17264]

x = np.arange (len(años))

width = 0.35
fig, ax = plt.subplots()

rects1 = ax.bar(x - width/2, aragon_radio, width, label='Aragon Radio')
rects2 = ax.bar(x + width/2, aragon_periodico, width, label='El periodico de Aragón')

ax.set_ylabel ("Euros")
ax.set_title ("Comparativa entre Aragon Radio (Azul) y El periodico de Aragon (Naranja)")
ax.set_xticks(x)
ax.set_xticklabels(años)

```

Vemos una gran diferencia en cuanto a inversión y dimensiones en los distintos medios
